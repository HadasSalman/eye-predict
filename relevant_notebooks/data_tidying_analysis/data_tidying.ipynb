# Data Tidying

```python
import numpy as np
import pandas as pd 
```

## Load parsed data into DF

```python
file_name = 'all_subjects_data_bmem_short_BDM'
df = pd.read_pickle('../parsing/' + file_name)
```

```python
df
```

## Find and remove Nan values

```python
#Select the data rows with Nan values
df[df.isnull().any(axis=1)]
#Remove Nan
tidyDf = df.dropna(how='any') 
```

## Find and remove "BLINK", "FIX", "SACC" data rows

```python
df.iloc[38:93]
```

```python
tidyDf = tidyDf[tidyDf['X_axis'].str.contains(' . ', na=False) == False]
tidyDf = tidyDf[(tidyDf['timeStamp'].str.contains('EFIX', na=False) == False) &
                        (tidyDf['timeStamp'].str.contains('EBLINK', na=False) == False) &
          (tidyDf['timeStamp'].str.contains('ESACC', na=False) == False)]
```

# Optional, set theblink data to '-10000'

```python
indexes = tidyDf[tidyDf['X_axis'].str.contains(' . ', na=False)].index
tidyDf.loc[indexes,'X_axis'] = '-10000.0'
tidyDf.loc[indexes,'Y_axis'] = '-10000.0'
```

## Change X, Y and timeStamp data from String to Numeric

```python
# Change X, Y and timeStamp data from String to Numeric
tidyDf.X_axis = pd.to_numeric(tidyDf.X_axis, errors='coerce')
tidyDf.Y_axis = pd.to_numeric(tidyDf.Y_axis, errors='coerce')
tidyDf.timeStamp = pd.to_numeric(tidyDf.timeStamp, errors='coerce')
```

## Save tidy data to pickle

```python
tidyDf.to_pickle('tidy_data_for_analysis')
```
